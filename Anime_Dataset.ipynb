```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/anime.csv")
```

```python
df["type"]=df["type"].str.lower()
df["type"]
```

```python
en_yuksek_puan_sayisi = df.nlargest(10, "rating")

plt.figure(figsize=(12,6))
plt.plot(en_yuksek_puan_sayisi["name"], en_yuksek_puan_sayisi["rating"], marker='o', linestyle='-', color='navy')
plt.xlabel('Anime Adı')
plt.ylabel('Puan Sayısı')
plt.title('En Yüksek Puana Sahip 10 Anime')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
en_yuksek_puan_sayisi = df.nlargest(9, "rating")

colors = ['lightblue', 'blue', 'navy', 'lightblue', 'blue', 'navy', 'lightblue', 'blue', 'navy']

plt.figure(figsize=(10, 6))
plt.pie(en_yuksek_puan_sayisi["rating"], labels=en_yuksek_puan_sayisi["name"], colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("En Yüksek Puana Sahip 9 Anime")
plt.axis('equal')
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(9, 6))
plt.bar(en_yuksek_puan_sayisi["name"], en_yuksek_puan_sayisi["rating"], color=colors)
plt.xlabel('Anime Adı')
plt.ylabel('Puan')
plt.title('En Yüksek Puana Sahip 9 Anime')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
en_dusuk_puan_sayisi = df.nsmallest(5, "rating")

# Sütun grafiği çiz
plt.figure(figsize=(5, 6))
plt.bar(en_dusuk_puan_sayisi["name"], en_dusuk_puan_sayisi["rating"], color=colors)
plt.xlabel('Anime Adı')
plt.ylabel('Puan')
plt.title('En Düşük Puana Sahip 5 Anime')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
en_yuksek_members = df.nlargest(10, "members")

# Sütun grafiği çiz
plt.figure(figsize=(6, 6))
plt.bar(en_yuksek_members["name"], en_yuksek_members["members"], color=colors)
plt.xlabel('Anime Adı')
plt.ylabel(' Dağılım')
plt.title('En Yüksek Fan Kitlesine Sahip 10 Anime')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
en_dusuk_members = df.nsmallest(10, "members")

plt.figure(figsize=(9, 10))
plt.bar(en_dusuk_members["name"], en_dusuk_members["members"], color=colors)
plt.xlabel('Anime Adı')
plt.ylabel(' Dağılım')
plt.title('En Düşük Fan Kitlesine Sahip 10 Anime')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
en_yuksek_members = df.nlargest(9, "members")

plt.figure(figsize=(10, 8))
plt.pie(en_yuksek_members["members"], labels=en_yuksek_members["name"], colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("En Yüksek Fan Kitlesine Sahip 10 Anime")
plt.axis('equal')
plt.tight_layout()
plt.show()
```
